```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
kalpkriz = pd.read_csv("https://raw.githubusercontent.com/vektorelpython20/Selinsbot/master/Ediz/heart-disease.csv")
len(kalpkriz)
kalpkriz.head()
```

```python
kalpkriz["age"].plot.hist(bins=50);
```

```python
kalpkriz.plot.hist(figsize=(10,30),subplots=True);
```

```python
Yas50 = kalpkriz[kalpkriz["age"]>50]
Yas50
```

```python
Yas50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,6));
```

```python
Yas50 = kalpkriz[kalpkriz["age"]>50]
fig,ax = plt.subplots(figsize=(10,6))
Yas50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,6),ax=ax);
ax.set_xlim([45,100]);
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(Yas50["age"],Yas50["chol"],c=Yas50["target"])
ax.set(title="Kalp Hastal ve Kolesterol Seviyeleri ",xlabel="Ya",ylabel="Kolesterol");
ax.legend(*scatter.legend_elements(),title="Haslatk Var/Yok");
```

```python

fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,sharex = True,figsize=(10,8))

scatter = ax0.scatter(Yas50["age"],Yas50["chol"],c=Yas50["target"],cmap="summer")
ax0.set(title="Kalp Hastal ve Kolesterol Seviyeleri ",ylabel="Kolesterol")
ax0.legend(*scatter.legend_elements(),title="Haslatk Var/Yok");

ax0.axhline(y=Yas50["chol"].mean(),color="r",linestyle="--",label="Averaj")

scatter = ax1.scatter(Yas50["age"],Yas50["thalach"],c=Yas50["target"],cmap="summer")
ax1.set(title="Kalp Hastal ve Max Kalp At Seviyesi",ylabel="Maksimum Kalp At",xlabel="Ya")
ax1.legend(*scatter.legend_elements(),title="Haslatk Var/Yok");

ax1.axhline(y=Yas50["thalach"].mean(),color="r",linestyle="--",label="Averaj")

fig.suptitle("Kalp Hastal Analizi",fontsize=16,fontweight='bold');
```
